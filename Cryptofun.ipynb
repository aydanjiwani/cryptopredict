```python
import cryptowatch as cw
from datetime import datetime, timedelta

assets = cw.assets.list()
```

```python
kraken = cw.markets.list("kraken")
#getting data and cleaning it
```

```python
candles = []
for market in kraken.markets:
  if(market.pair.endswith("usd")):
    print(market.pair)
    ticker = "{}:{}".format(market.exchange, market.pair).upper()
    candles.append(cw.markets.get(ticker, ohlc=True, periods=["4h"]))

```

```python
usdmarkets = []
for market in kraken.markets:
  if(market.pair.endswith("usd")):
    usdmarkets.append(market)
```

```python
for i in range(0,len(candles)):
  print(usdmarkets[i].pair + " " + str(candles[i].of_4h[-1][1]))
```

```python
timeseries = {}
for i in range(0,len(candles)):
  timeseries[usdmarkets[i].pair] = candles[i].of_4h
print(timeseries["btcusd"])
```

```python
import pandas as pd
```

```python
dataframes = {}
for i in timeseries:
  dataframes[i] = pd.DataFrame(timeseries[i], columns=['Date','Open','Close','High','Low','Vol','Quotevol'])
```

```python
from datetime import datetime
for i in dataframes:
  dataframes[i]['Date'] = pd.to_datetime(dataframes[i]['Date'], unit='s')
```

```python
print(dataframes["btcusd"])
```

```python
#from here on, building an LSTM and using the btcusd dataframe to test optimal param set
df = dataframes["btcusd"]
training_set = df.iloc[:800, 1:2].values #training set will be first 800 values
test_set = df.iloc[800:, 1:2].values #training set will be most recent 200 values, approx last 33 days
```

```python
import math
import matplotlib.pyplot as plt
import keras
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import *
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from keras.callbacks import EarlyStopping

# Feature Scaling
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)# Creating a data structure with 60 time-steps and 1 output
X_train = []
y_train = []
for i in range(60, 800):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
model = Sequential()#Adding the first LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
model.add(Dropout(0.2))

# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))

# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))

# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))

# Adding a fourth LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50))
model.add(Dropout(0.2))

model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')

model.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
#converting above code to a pipeline with params
def pipeline(coin,delay,neurons,layers,epochs,batchsize): #min 2 for layers, coin must be in prev set
  df = dataframes[coin+"usd"]
  training_set = df.iloc[:800, 1:2].values #training set will be first 800 values
  test_set = df.iloc[800:, 1:2].values #training set will be most recent 200 values, approx last 33 days
  sc = MinMaxScaler(feature_range = (0, 1))

  training_set_scaled = sc.fit_transform(training_set)# Creating a data structure with 60 time-steps and 1 output
  X_train = []
  y_train = []
  for i in range(delay, 800):
      X_train.append(training_set_scaled[i-delay:i, 0])
      y_train.append(training_set_scaled[i, 0])
  X_train, y_train = np.array(X_train), np.array(y_train)
  X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

  #model building

  model = Sequential()#Adding the first LSTM layer and some Dropout regularisation
  model.add(LSTM(units = neurons, return_sequences = True, input_shape = (X_train.shape[1], 1)))
  model.add(Dropout(0.2))

  for i in range(0,layers-2):
    model.add(LSTM(units = neurons, return_sequences = True))
    model.add(Dropout(0.2))


  # Adding a fourth LSTM layer and some Dropout regularisation
  model.add(LSTM(units = neurons))
  model.add(Dropout(0.2))

  model.add(Dense(units = 1))

  model.compile(optimizer = 'adam', loss = 'mean_squared_error')

  model.fit(X_train, y_train, epochs = epochs, batch_size = batchsize)

  #data prediction
  dataset_train = df.iloc[:800, 1:2]
  dataset_test = df.iloc[800:, 1:2]
  dataset_total = pd.concat((dataset_train, dataset_test), axis = 0)
  inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values #note: we need the first 60 inputs of train data in test data
  inputs = inputs.reshape(-1,1)
  inputs = sc.transform(inputs)
  X_test = []
  for i in range(delay, 200+delay):
      X_test.append(inputs[i-delay:i, 0])
  X_test = np.array(X_test)



  X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
  print(X_test.shape)

  predicted_stock_price = model.predict(X_test)
  predicted_stock_price = sc.inverse_transform(predicted_stock_price)

  plt.plot(df.loc[800:, 'Date'],dataset_test.values, color = 'red', label = 'Real Price')
  plt.plot(df.loc[800:, 'Date'],predicted_stock_price, color = 'blue', label = 'Predicted Price')
  plt.gcf().autofmt_xdate()
  plt.title('Real vs Predicted ' + coin + ' price')
  plt.xlabel('Date')
  plt.ylabel('Price (USD)')
  plt.legend()
  plt.show()

  mse = tf.keras.losses.MeanSquaredError()
  mae = tf.keras.losses.MeanAbsoluteError()
  print("Current coin: " + coin)
  print("Mean Squared Error: " + str(mse(dataset_test.values, predicted_stock_price).numpy()))
  print("Mean Absolute Error: " + str(mae(dataset_test.values, predicted_stock_price).numpy()))
```
